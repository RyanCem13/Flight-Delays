```python
#Identify how many missing values exist
delay.isnull().sum()
```

```python
#Check for null values
null_values = delay.isnull().any(axis=1) 
```

```python
#checking to see how relevant the missing values are to know how to deal with them 
delay[null_values]
```

```python
#This is a very small amount of the entire dataset and not able to be salvaged, we will drop these values
delay.dropna(inplace=True)
```

```python
#Check missing values were dropped
delay.isnull().sum()
```

```python
#Recheck shape of data
delay.shape
```

```python
#Create new dataframe of duplicates
df_dups = delay[delay.duplicated()]
```
